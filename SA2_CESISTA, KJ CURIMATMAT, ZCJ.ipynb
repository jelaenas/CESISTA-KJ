## I. Forest fires are a serious problem for the preservation of the Tropical Forests. Understanding the frequency of forest fires in a time series can help to take action to prevent them.

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\cjere\Downloads\amazon.csv',  encoding='latin1')

month_translation = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'Março': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}

#1
print("Top 10 rows of the dataset:")
print(df.head(10))

#2
print("\nLast 10 rows of the dataset:")
print(df.tail(10))

#3
print("\nShape of the dataset(num_rows, num_columns):")
print(df.shape)

#4
print("\nInformation about the dataset:")
print(df.info())

#5
duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
    print("\nDuplicate rows:")
    print(duplicate_rows)

df = df.drop_duplicates()

#6
print("\nNull values in the dataset:")
print(df.isnull().sum())

#7
print("\nOverall statistics about the dataframe:")
print(df.describe())

#8
df['month'] = df['month'].map(month_translation)

#9
total_fires = df['number'].sum()
print("\nTotal number of fires registered:", total_fires)

#10
df = df.dropna(subset=['month'])

if not df.empty:
    max_fires_month = df.groupby('month')['number'].sum().idxmax()
    print("\nMonth with maximum number of forest fires reported:", max_fires_month)
else:
    print("\nThere are no valid entries for the month.")

#11
try:
    max_fires_year = df.groupby('year')['number'].sum().idxmax()
    print("\nYear with maximum number of forest fires reported:", max_fires_year)
except ValueError:
    print("\nThere are no valid entries for the year.")

#12
try:
    max_fires_state = df.groupby('state')['number'].sum().idxmax()
    print("\nState with maximum number of forest fires reported:", max_fires_state)
except ValueError:
    print("\nThere are no valid entries for the state.")

#13
total_fires_amazonas = df[df['state'] == 'Amazonas']['number'].sum()
print("\nTotal number of fires reported in Amazonas:", total_fires_amazonas)

#14
print("\nNumber of fires reported in Amazonas (year-wise):")
print(df[df['state'] == 'Amazonas'].groupby('year')['number'].sum())

#15
print("\nNumber of fires reported in Amazonas (day-wise):")
print(df[df['state'] == 'Amazonas'].groupby('date')['number'].sum())

#16
total_fires_2015 = df[df['year'] == 2015]['number'].sum()
print("\nTotal number of fires reported in 2015:", total_fires_2015)
print("\nVisualization of fires reported in 2015 (month-wise):")
print(df[df['year'] == 2015].groupby('month')['number'].sum())

#17
print("\nAverage number of fires reported from highest to lowest (state-wise):")
print(df.groupby('state')['number'].mean().sort_values(ascending=False))

#18
states_dec = df[df['month'] == 'December']['state'].unique()
print("\nStates where fires were reported in 'December' month:", states_dec)
```
