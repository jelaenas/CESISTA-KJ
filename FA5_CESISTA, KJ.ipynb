```python
import turtle
import math
import random
import numpy as np


x = np.arange(5,8)
x = np.concatenate([x, -x])

y = np.arange(5,8)
y = np.concatenate([y, -y])
```

```python


def point_in(x,y, circle_x, circle_y, radius):
    distance = math.sqrt((x- circle_x) ** 2 + (y - circle_y) ** 2)
    return distance < radius


def draw_point(x, y, circle_x, circle_y, radius):
    if point_in(x, y, circle_x, circle_y, radius):
        turtle.penup()
        turtle.goto(x,y)
        turtle.dot(1, "blue")
        return True
    else:
        return False

radius = 150

# move
turtle.penup()
turtle.right(90)
turtle.forward(radius)
turtle.left(90)
turtle.pendown()

# circle
turtle.color("pink")
turtle.width(4)
turtle.circle(radius)


# move back
turtle.color("blue")
turtle.shape("turtle")
turtle.penup()
turtle.right(-90)
turtle.forward(radius)
turtle.left(-90)
turtle.pendown()

circle_x, circle_y = turtle.pos()

direction_x = 7
direction_y = 7

dirs_x = x
dirs_y = y
while True:
    x = turtle.xcor() 
    y = turtle.ycor()
    if(not draw_point(x + direction_x,y + direction_y, circle_x, circle_y, radius)):
        while(not draw_point(x + direction_x,y + direction_y, circle_x, circle_y, radius)):
            direction_x = random.choice(dirs_x)
            direction_y = random.choice(dirs_y)
        

turtle.done()
```
